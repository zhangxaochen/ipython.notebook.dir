# 使用堆实现Dijkstra算法

* [heapq模块](http://docs.python.org/2/library/heapq.html)

```python
import heapq

from PIL import Image
from IPython import display as disp
from io import BytesIO

def show(im):
    b = BytesIO()
    im.save(b, format='png')
    disp.display(disp.Image(data=b.getvalue(), format='png', embed=True))

darkmaze = Image.open(r'D:\Documents\Desktop\darkmaze.jpg')
im = darkmaze.convert('L').point(lambda x: 0 if x < 200 else 1, '1')

def dark(p):
    '''p点周围的7x7邻域内只要包含一处黑色的墙则认为没有灯'''
    def inner():
        for ox in range(-3, 4):
            for oy in range(-3, 4):
                x, y = p
                yield darkmaze.getpixel((x + ox, y + oy)) == (0, 0, 0)
    return max(inner())
def memoed(f):
    memo = {}
    def inner(*args):
        if args not in memo:
            memo[args] = f(*args)
        return memo[args]
    return inner

@memoed
def cost(p):
    '''没灯的地方举步维艰...'''
    return 1000 if dark(p) else 1

start = (402, 984)
end = (398, 24)
top, right, bottom, left = 22, 793, 986, 7

def forbidden(p):
    x, y = p
    return x < left or y < top or x > right or y > bottom or im.getpixel(p) == 0

def dij():
    prev = {start: None}
    next = [(0 ,start)]
    closed = {}
    d = {start: 0}
    
    
    while next:
        # index = min(range(len(next)), key=lambda i: d[next[i]]) 因为heappop自动删除最小项
        
        num, source = heapq.heappop(next)    # source = next.pop(index)。用unpack来代替
        
        if source == end:
            break
        if source in closed:
            continue
        closed[source] = 1
        
        for ox, oy in ((0, 1), (1, 0), (0, -1), (-1, 0)):
            x, y = source
            target = (x + ox, y + oy)
            if not forbidden(target) and target not in closed:
                if d[source] + cost(target) < d.get(target, float('inf')):
                    d[target] = d[source] + cost(target)
                    prev[target] = source
                    heapq.heappush(next, (d[target], target))
            
    path = []
    p = end
    while p in prev:
        path.append(p)
        p = prev[p]
    path = list(reversed(path))
    
    out = darkmaze.point(lambda p: p / 2).convert('RGB')
    
    for p in path:
        out.putpixel(p, (255, 255, 255))
    
    show(out)
dij()
```
