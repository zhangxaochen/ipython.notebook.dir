```python
import urllib2
from urlparse import urljoin
import re
import os
import sys
import hashlib

'boardID:'
bid=509
page=1
base_url=r'http://acm.zju.edu.cn'
# url=r'http://www.cc98.org/list.asp?boardid=%d&page=%d'%(bid, page)
# r=urllib2.urlopen(url)
# c=r.read()

def fetch_page(url):
    ur'根据 url 获取网页或本地缓存'
    
    cache_dir = 'cache'
    if not os.path.isdir(cache_dir):
        os.makedirs(cache_dir)
    
    fn = hashlib.md5(url).hexdigest()
    rel_path = os.path.join(cache_dir, fn)
    '若未缓存过：'
    if not os.path.isfile(rel_path):
        with open(rel_path, 'w') as f:
            try:
                r = urllib2.urlopen(url)
                f.write(r.read())
            except URLError, e:
                print '---------------:', url
                print e
    return open(rel_path).read()


def get_pages():
    ur'获得所有列表页面的 url'
    content = fetch_page(r'http://acm.zju.edu.cn/onlinejudge/showProblems.do?contestId=1')
    return [urljoin(base_url, i) 
            for i in sorted(set(re.findall(r'/onlinejudge/showProblems\.do\?contestId=1&pageNumber=\d+', content)),
                            key=lambda url: int(url.split('=')[-1]))]

def get_posts():
    u'获得所有题目的 url'
    posts=[]
    for url in get_pages():
        content = fetch_page(url)
        posts.extend(urljoin(base_url, i) 
                     for i in sorted(set(re.findall(r'/onlinejudge/showProblem\.do\?problemCode=\d+', content))))
    return posts

def contains(url, ptn):
    content = fetch_page(url)
    return re.search(ptn, content) is not None
    
def grepZOJ(ptn):
    return [u for u in get_posts() if contains(u, ptn)]
        
    
# ---------------test
get_pages()
urls = get_posts()
print len(urls)

'发现少了 2866， 2999 两道题:'
for i in range(len(urls)-1):
    if int(urls[i+1].split('=')[-1])-int(urls[i].split('=')[-1]) != 1:
        print urls[i].split('=')[-1]
    
good_urls = grepZOJ('[Mm]atri(x|ces)')
# good_urls = grepZOJ('[Mm]atrixes')
print len(good_urls)
good_urls
```
