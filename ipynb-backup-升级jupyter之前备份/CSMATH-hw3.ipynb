```python
from matplotlib import pyplot
from numpy import *
import pylab as pl
from numpy.linalg import det
import numpy.matlib as ml
import random

labels=[]

def calc_J():
    sum = 0
    for i in xrange(N):
        sum += norm(X[i]-centers[labels[i]])
    return sum

def distmat(X, Y):
    n = len(X)
    m = len(Y)
    xx = ml.sum(X*X, axis=1)
    yy = ml.sum(Y*Y, axis=1)
    xy = ml.dot(X, Y.T)
    return tile(xx, (m, 1)).T+tile(yy, (n, 1)) - 2*xy

def init_params(centers,k):
    #使用kmeans算法设置EM算法的初始参数
    pMiu = centers
    pPi = zeros([1,k], dtype=float)
    pSigma = zeros([len(X[0]), len(X[0]), k], dtype=float)
    #计算X到其他点的距离
    dist = distmat(X, centers)
    #分配X到最近的中心
    labels = dist.argmin(axis=1)
    #重新计算每一个中心和参数
    for j in range(k):
        idx_j = (labels == j).nonzero()
        pMiu[j] = X[idx_j].mean(axis=0)
        pPi[0, j] = 1.0 * len(X[idx_j]) / N
        pSigma[:, :, j] = cov(mat(X[idx_j]).T)
    return pMiu, pPi, pSigma

def calc_prob(k,pMiu,pSigma):
    Px = zeros([N, k], dtype=float)
    for i in range(k):
        Xshift = mat(X - pMiu[i, :])
        inv_pSigma = mat(pSigma[:, :, i]).I
        coef = pow((2*pi), (len(X[0])/2)) * sqrt(det(mat(pSigma[:, :, i])))            
        for j in range(N):
            tmp = (Xshift[j, :] * inv_pSigma * Xshift[j, :].T)
            Px[j, i] = 1.0 / coef * exp(-0.5*tmp)
    return Px

#随机产生要聚类的数据
def gengauss(size):
    import numpy as np
    mean = [0,10]
    cov = [[1,0],[0,100]]
    x,y = np.random.multivariate_normal(mean,cov,size).T
    return x, y



def MoG(X, k, threshold=1e-15, maxiter=300):
    N = len(X)
    labels = zeros(N, dtype=int)
    centers = array(random.sample(X, k))
    iter = 0
    pMiu, pPi, pSigma = init_params(centers,k)
    Lprev = float('-10000')
    pre_esp = 100000
    while iter < 100:
        Px = calc_prob(k,pMiu,pSigma)
        #计算新的pGamma ([N, k])值
        pGamma =mat(array(Px) * array(pPi))
        pGamma = pGamma / pGamma.sum(axis=1)
        #计算新的pMiu, pPi, pSigma值
        Nk = pGamma.sum(axis=0) #[1, K]
        pMiu = diagflat(1/Nk) * pGamma.T * mat(X) # Miu=[K, 2]
        pPi = Nk / N #[1, K]
        pSigma = zeros([len(X[0]), len(X[0]), k], dtype=float)
        for j in range(k):
            Xshift = mat(X) - pMiu[j, :]
            for i in range(N):
                pSigmaK = Xshift[i, :].T * Xshift[i, :]
                pSigmaK = pSigmaK * pGamma[i, j] / Nk[0, j]
                pSigma[:, :, j] = pSigma[:, :, j] + pSigmaK
        #更新labels
        labels = pGamma.argmax(axis=1)
        iter = iter + 1
        #检查是否收敛
        L = sum(log(mat(Px) * mat(pPi).T))
        cur_esp = L-Lprev
        if cur_esp < threshold:
            break
        if cur_esp > pre_esp:
            break
        pre_esp=cur_esp
        Lprev = L
     #   print "iter %d esp %lf" % (iter,cur_esp)
#     print labels
    return labels

if __name__ == '__main__':
    #产生样本
    samples = gengauss(1000)
    N = len(samples[0])
    X = zeros((N, 2))
    for i in range(N):
        X[i, 0] = samples[0][i]
        X[i, 1] = samples[1][i]
    pl.plot(X[:,0],X[:,1],'g.');
    classid=MoG(X, 3);
 #   print classid
    #不同的类绘制出不同的颜色来表示
    for i in range(N):
        if classid[i][0]==0:
           pl.plot(X[i,0],X[i,1],'r.');
        elif classid[i,0]==[1]:
            pl.plot(X[i,0],X[i,1],'g.');
        elif classid[i,0]==[2]:
            pl.plot(X[i,0],X[i,1],'b.');
        else:
            pl.plot(X[i,0],X[i,1],'y.');
        

#     pl.show()
```
