```python
import sys
import pylab as pl
import numpy as np

Mat = []
Matrix = []
countOfThree = 0 

#从下载的手写字体中是“3”的取出来
def ReadDigitalThree(filePath):
    file = open(filePath)
    Xi = []  
    
    while 1:
        lines = file.readlines(64000)   
        if not lines:
            break
        
        for line in lines:
            line=line.strip('\n')
              
            if len(line) < 5 :
                number = int(line)
                if  number == 3:
                    #save the vector to the matrix
                    Mat.append(Xi)  
                Xi = []
            else:
                for str in line:
                    Xi.append(int(str))

#创建kmeans类对PCA结果进行简单聚类
class kmeans:
	def __init__(self,k,data):

		self.nData = np.shape(data)[0]
		self.nDim = np.shape(data)[1]
		self.k = k
		
	def kmeanstrain(self,data,maxIterations=50):
		
		print "\nData Amount:", self.nData
		print "\nData Dimension:", self.nDim

             	# 找每个特征中的最大最小值
		minima = min([v for k,v in data])
		maxima = max([v for k,v in data])

		
		self.centres = np.random.rand(self.k,self.nDim)*(maxima-minima)+minima
		oldCentres = np.random.rand(self.k,self.nDim)*(maxima-minima)+minima
	
		count = 0
		#print centres
		while np.sum(np.sum(oldCentres-self.centres))!= 0 and count<maxIterations:
	
			oldCentres = self.centres.copy()
			count += 1
	
			# 计算距离
			distances = np.ones((1,self.nData))*np.sum((data-self.centres[0,:])**2,axis=1)
			for j in range(self.k-1):
				distances = np.append(distances,np.ones((1,self.nData))*np.sum((data-self.centres[j+1,:])**2,axis=1),axis=0)
	
			# 确定属于哪个类
			cluster = distances.argmin(axis=0)
			cluster = np.transpose(cluster*np.ones((1,self.nData)))
	
			# 更新每个类的中心	
			for j in range(self.k):
				thisCluster = np.where(cluster==j,1,0)
				if np.sum(thisCluster)>0:
					self.centres[j,:] = np.sum(data*thisCluster,axis=0)/np.sum(thisCluster)
	
		return self.centres
	
	def kmeansfwd(self,data):
		
		nData = np.shape(data)[0]
		# 计算距离
		distances = np.ones((1,nData))*np.sum((data-self.centres[0,:])**2,axis=1)
		for j in range(self.k-1):
			distances = np.append(distances,np.ones((1,nData))*np.sum((data-self.centres[j+1,:])**2,axis=1),axis=0)
	
		# 分类
		cluster = distances.argmin(axis=0)
		cluster = np.transpose(cluster*np.ones((1,nData)))
	
		return cluster
    
if __name__ == '__main__':  
    fname=r'D:\Documents\Desktop\optdigits-orig.tra'
    ReadDigitalThree(fname)
    
    #将手写“3”数据存储
    Matrix = np.matrix(Mat).T
    
    #得到中心矩阵
    MatrixCenter = Matrix - Matrix.mean(1)
    print MatrixCenter
    # 运行svd
    U, s, V = np.linalg.svd(MatrixCenter, full_matrices=True)
      
    # 将svd结果U的前两列取出作为最大的特征
    Matrix_from_U = np.matrix(np.array([(U[:,0]), (U[:,1])])).T

    Matrix_lenda = Matrix_from_U.getI()* MatrixCenter
    #用于lambda显示
    x_show=np.matrix(np.array(Matrix_lenda[0,:]))
    y_show=np.matrix(np.array(Matrix_lenda[1,:]))

    pl.plot(x_show, y_show, 'g.')
    k=16
    n=2
    
    data=Matrix_lenda.T.getA()

    #print "\ndata:",data
    
    #聚类
    KMS=kmeans(k,data)
    centres = KMS.kmeanstrain(data)
    clusters = KMS.kmeansfwd(data)
    #clusters = kmeans(data, k, threshold)
    #print "centres:",centres
    
    xc_show=np.matrix(np.array(centres[:,0]))
    yc_show=np.matrix(np.array(centres[:,1]))
    pl.plot(xc_show, yc_show, 'ro')
    pl.grid()
#     pl.show()
```
