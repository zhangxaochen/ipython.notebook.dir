```python
from sklearn.metrics import precision_recall_curve, auc
import pylab as pb
import joblib


def calc_prc_values(true, prob):
    precision, recall, thresholds = precision_recall_curve(true, prob)
    area = auc(recall, precision)
    return dict(precision=precision, recall=recall, thresholds=thresholds, area=area)


def plot_prc_one(d):
    pb.plot(d['recall'], d['precision'], label='area = %0.2f' % d['area'])
    pb.plot([0, 1], [0, 1], 'k--')
    
    
def plot_prc_common():
    gcf().set_size_inches(12, 12)
    pb.xlim([0.0, 1.0])
    pb.ylim([0.0, 1.0])
    pb.xlabel('recall')
    pb.ylabel('precision')
    pb.legend(loc="lower right")
    

true = joblib.load('cache/true')
prob = joblib.load('cache/prob')
d = calc_prc_values(true, prob)
plot_prc_one(d)
plot_prc_common()
pb.show()
```

```python
precision = d['precision']
recall = d['recall']
f1 = 2 * (precision * recall) / (precision + recall)

mask = np.isfinite(f1)
best_index = np.where(mask)[0][np.argmax(f1[mask])]
best_threshold = d['thresholds'][best_index]
print(best_threshold, f1[best_index], precision[best_index], recall[best_index])
```

```python
from sklearn.metrics import classification_report


print(classification_report(true, prob > best_threshold))
```
