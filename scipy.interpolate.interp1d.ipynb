```python
from scipy.interpolate import interp1d
from pylab import *
```

```python
x=np.arange(10)
#y=[0,1]*5
y=np.random.randint(10, size=10)
plot(x, y)
show()
```

```python
f=interp1d(x, y, kind='cubic')
f2=interp1d(x, y, kind=2)
#nx=np.arange(0, 9, .05)
nx=linspace(0, 9, 333)
plot(nx, f(nx),'b', x, y, 'or')
plot(nx, f2(nx), 'g')
show()
```

```python
size=100
x=arange(size)
y=np.random.randint(20, size=size)
f=interp1d(x, y, kind='cubic')
nx=linspace(0, 99, 1000)
plot(x, y,'r', nx, f(nx), 'b')
show()
```

```python
x=arange(10)
y1=[0,1]*5
y2=[0,1]*4+[0, 222]
f1=interp1d(x, y1, kind=3)
f2=interp1d(x, y2, kind=3)
nx=linspace(0, 9, 100)
plot(x, y1, nx, f1(nx), nx, f2(nx))
show()
```

```python
#尝试分段三次样条
y=random(15)

left1=0
right1=10
x1=arange(left1, right1)
y1=y[:right1-left1]
f1=interp1d(x1, y1, 3)
nx1=linspace(left1, right1-1, 100)

left2=1
right2=15
x2=arange(left2, right2)
y2=y[left2:right2]
f2=interp1d(x2, y2, 3)
nx2=linspace(left2, right2-1, 100)
plot(x1, y1, nx1, f1(nx1), nx2, f2(nx2))
show()
```
